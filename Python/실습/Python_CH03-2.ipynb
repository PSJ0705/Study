```python
from pickle import ADDITEMS

from IPython.core.tbtools import count_lines_in_py_file

# 열번 찍어 안 넘어가는 나무 없다, 라는 속담을 파이썬 프로그램을 만들면 다음과 같다

hit = 0

while hit <10:
    hit += 1
    print("나무를 %d번 찍었습니다." %hit)
    if hit == 10:
        print("나무 넘어갑니다.")
```

```python
# 위 예에서 while의 조건문은 hit < 10이다. 즉, hit이 10보다 작은 동안 while 문에 포함된 문장들을 계속 수행한다.
```

```python
# while문 만들기
# 여러가지 선택지 중 하나를 선택해서 입력받는 예제를 만들기

prompt = """
    1. Add
    2. Del
    3. List
    4. Quit
    Enter Number:"""

number = 0

while number != 4:
    print(prompt)
    number = int(input())       #number = int(input())은 사용자에게 숫자 입력을 받아들임
```

While문 강제로 빠져나가기 (break)

```python
# 커피 자판기 이야기를 파이썬 프로그램

coffe = 10
money = 300

while money:
    print("돈을 받았으니 커피를 줍니다.")
    coffe = coffe -1
    print("남은 커피의 양은 %d 입니다." % coffe)
    if coffe == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다. 잔여 커피 수량 : %d" %coffe)
        break
```

```python

coffe = 10

while True :
    money = int(input("돈을 넣어주세요"))
    if money == 300:
        print(" 커피를 줍니다")
        coffe = coffe -1
    elif money > 300:
        print("거스름돈 %d를 주고 커피를 줍니다." %(money - 300))
        coffe = coffe -1
    else:
        print("%d원을 돌려주고 커피를 주지 않습니다." %money)
        print("%d원만큼 돈이 부족합니다." %(300 - money))
        print("남은 커피의 양은 %d개 입니다." %coffe)
    if coffe == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
        break
```

```python
# 1부터 10까지의 숫자 중에서 홀수만 출력하는 것을 while 문을 사용해서 작성하는 방버

a = 0
while a<10:
    a += 1
    if a % 2 == 0: continue     # a가 짝수면 continue문을 수행해서 while의 맨 처음으로 돌아가기 때문에 print(a)가 수행되지 않음
    print(a)
```

While-else 문

While-else 문은 While 문이 정상적으로 종료되었을 때(break로 빠져나가지 않을 때) else 절이 실행된다.

```python
count = 0

while count <= 5:
    print("카운트 : %d" %count)
    count += 1
else:
    print("while문이 정상 종료되었습니다.")
```

```python
# 하지만 break 문으로 while 문을 빠져나가면 else 절은 실행되지 않는다.

count = 0

while count < 5:
    print("카운트 : %d" %count)
    count += 1
    if count == 2:
        print("while문이 break로 종료되었습니다.")
        break
    else:
        print("while문이 정상적으로 종료되었습니다.")
```

중첩된 while 문

```python
#while문 안에 또 다른 while문을 사용할 수 있다.

i = 1

while i <=3:
    j = 1
    while j <= 3:
        print(f"i={i}, j={j}")
        j += 1
    i += 1

# 중첩된 while 문에서 break나 continue를 사용할 때는 가장 가까운 while 문에만 영향을 준다.
```
