AND, OR, NOT

    x or y : x 와 y 중 하나만 참이어도 참이다.
    x and y : x 와 y 모두 참이어야 한다.
    not x :  x가 거짓이면 참이다.

```python
# 돈이 3000원 이상 있거나 카드가 있다면 택시를 타고, 그렇지 않다면 걸어가라

money = 2000
Card = True

if money >= 3000 or Card:       #card = True 이기 때문에 참이 된다.
    print("택시를 타라")
else:
    print("걸어가라")
```

In, Not In

    In : x in 리스트, x in 튜플, x in 문자열
    Not in : x not in 리스트, x not in 튜플, x not in 문자열

```python
print(1 in [1,2,3])         #[1,2,3]이라는 리스트 안에 1이 있는가? 라는 조건문
print(4 not in [1,2,3])     #[1,2,3]이라는 리스트 안에 4가 없는가? 라는 조건문
print(4 in [1,2,3])
```

```python
# 만약 주머니에 돈이 있으면 택시를 타고, 없으면 걸어가라

poket = ['paper', 'money', 'cellphone']

if 'money' in poket:
    print("택시를 타라")
else:
    print("걸어가라")
```

```python
# 조건문에서 아무 일도 하지 않게 설정하는 방법

# 주머니에 돈이 있다면 가만히 있고, 주머니에 돈이 없다면 카드를 꺼내라.

pocket = ['paper', 'money', 'cellphone']

if 'money' in pocket:
    pass        #pocket 리스트 안에 money 문자열이 있기 때문에 pass 가 수행되고 아무런 결괏값도 보여주지 않는다.
else:
    print("카드를 꺼내라.")
```

다양한 조건을 판단하는 elif

```python
# 주머니에 돈이 있으면 택시를 타고, 주머니에 돈은 없지만 카드가 있다면 택시를 타고, 돈도 없고 카드도 없다면 걸어가라.

pocket = ['paper', 'cellphone']
card = True

if 'money' in pocket:
    print("택시를 타고 가라")
elif card:          # elif는 이전 조건문이 거짓일때 수행된다.
    print("택시를 타고 가라.")
else:
    print("걸어가라")
```

비교 연산자의 연쇄 작용

```python
# 파이썬에서는 수학에서처럼 비교 연산자를 연쇄적으로 사용할 수 있다.

x = 5

print(1 < x < 10)           # x가 1보다 크고 10보다 작은가?

print(10 <= x <= 20)        # x가 10 이상 20 이하인가?
```

```python
# 위 코드는 다음과 같이 쓰는 것과 같다

x = 5

print((1<x) and (x < 10))

print((10<=x) and (x <= 20))
```

조건부 표현식

```python
# 점수가 60점 이상이면 "합격", 미만이면 "불합격"이라는 메시지를 저장하는 코드

score = 85
result = "합격" if score>=60 else "불합격"
# 변수 = 참일_때_값   if 조건 else  거짓일_때_값

print(result)

```
