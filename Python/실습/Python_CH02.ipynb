format 함수를 사용한 포매팅

```python
print("I eat {0} apples".format(3))
print("I eat {0} apples" .format("five"))
```

```python
#숫자값을 가진 변수로 대입하기
number = 10
print("I eat {0} apples" .format(number))
```

```python
#2개 이상의 값 넣기

number = 123
day = "three"
print("I ate {0} apples. so I was sick for {1} days." .format(number, day))

#문자열의 {0}, {1}과 같은 인덱스 항목이 format 함수의 입력값으로 순서에 맞게 바뀐다.
```

```python
#이름으로 넣기

print("I ate {number} apples. so I was sick for {day} days." .format(number = 10, day = 123))

#{name} 형태를 사용할 경우 반드시 name=value와 같은 형태의 입력값이 있어야 한다.
```

```python
#인덱스와 이름을 혼용해서 넣기

print("I ate {0} apples. so I was sick for {day} days." .format(10, day = 3))
```

```python
#왼쪽 정렬
print("{0:<10}".format("hi"))

#오른쪽 정렬 : 화살표 방향을 생각하면 쉬움
print("{0:>10}" .format("hi"))

#가운데 정렬
print("{0:^10}".format("hi"))

#공백 채우기
print("{0:=^10}".format("hi"))
print("{0:~<10}".format("hi"))
```

```python
#소수점 표현하기
y = 3.42134234
print("{0:0.4f}".format(y))

print("{0:10.4f}".format(y))
```

```python
#{또는} 문자 표현하기

print("{{ and }} ".format())
```

f 문자열 포매팅

```python
name = '홍길동'
age = 29
f'나의 이름은 {name} 입니다. 나이는 {age}입니다.'

#f 문자열 포매팅은 변숫값을 생성한 후, 그 값을 참조할 수 있다.
```

```python
age = 30
f'나는 내년이면 {age+1}살이 된다.'

#f 문자열 포매팅은 표현식을 지원한다.
#표현식이란, 중괄호 안의 변수를 계산식과 함께 사용하는 것을 말한다.
```

```python
#f 문자열에서 딕셔너리

d = {'name' : '홍길동', 'age' : 30}
f'나의 이름은 {d["name"]}입니다. 나이는 {d["age"]}입니다.'
```

```python
#정렬

f'{"hi":<10}' #왼쪽 정렬

f'{"hi":>10}' #오른쪽 정렬

f'{"hi":^10}' #가운데 정렬
```

```python
#공백 채우기

f'{"hi":=^10}'
f'{"hi":=>10}'
f'{"hi":=<10}'
```

```python
#소수점 표현하기

y = 3.421342324
f'{y:0.4f}' #소수점 4자리까지만 표현
f'{y:10.4}' #소수점 4자리까지 표현하고 총 자리수를 10으로 맞춤
```

```python
#문자열 바꾸기 - replace

a = "Life is too short"
a.replace("Life", "Your leg")

#replace함수는 replace(바뀔 문자열, 바꿀 문자열)처럼 사용해서 문자열 안의 특정한 값을 다른 값으로 치환해준다.
```

```python
#문자열 나누기 - split

a = "Life is too short"
print(a.split())

#split 함수는 ()안에 아무것도 넣어주지 않으면 공백을 기준으로 문자열을 나누어 준다.


b = "a:b:c:d"
print(b.split(':'))

#b.split(":")처럼 괄호 안에 특정 값이 있을 경우에는 괄호 안의 값을 구분자로 해서 문자열을 나누어준다.
```

```python
#문자열이 알파벳으로만 구성되어 있는지 확인하기 - isalpha

s = "Python"
print(s.isalpha())

s = "Python3"
print(s.isalpha())

s = "Hello World"
print(s.isalpha())

# 공백이나 숫자, 특수문자가 포함되어 있으면 False를 반환함
```

```python
#문자열이 숫자로만 구성되어 있는지 확인하기  - isdigit

s = "12345"
print(s.isdigit())

s = "123445a"
print(s.isdigit())

s = "123 123123 123"
print(s.isdigit())
```

```python
#문자열이 특정 문자열로 시작하는지 확인하기 - startswith

s = "Life is too short"

print(s.startswith("Life"))

print(s.startswith("short"))

#startswith 함수는 문자열이 지정한 문자(열)로 시작하는지를 검사한다.
```

```python
#문자열이 특정 문자(열)로 끝나는지 확인하기 - endswith

s = "Life is too short"

print(s.endswith("short"))

print(s.endswith("Life"))
```
