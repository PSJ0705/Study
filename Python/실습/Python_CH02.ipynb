8진수와 16진수

```python
#8진수, 0o로 시작함
a = 0o177
print(a)
```

```python
#8진수와 10진수를 더하면?
b = 13
print(a+b)
```

```python
#16진수,0x로 시작
a = 0x8ff
b = 0xABC
print(a+b)
```

```python
#10진수를 8, 16진수로 편하게 바꾸거나/출력하는 방법은?
#내장함수 oct()를 사용하면 됨
a = 50
#a를 8진수로 변환하기
def oct(a):
    return format(a, '08') #8자리 폭으로 0을 채워라
print(oct(a))
#10진수가 10진수로 출력됨, 왜?
```

```python
print(oct(a))

a = 50

print(oct(a))
print(format(a, 'o'))
```

```python

#10진수로 변환해주는 함수
#init()

#2진수로 변환해주는 함수
#bin()

#8진수로 변환해주는 함수
#oct()

#16진수로 변환해주는 함수
#hex()
```

```python

bin(5)
print(bin(5))

oct(5)
print(oct(5))
```

```python
#진수 바꿔서 출력하기
a = 1534
print(format(a, 'o')) #o는 8진수(Octal)을 의미
print(format(a, 'b')) #b는 2진수를 의미
print(format(a, 'x')) #x는 16진수 의미
```
