```python
# 딕셔너리에는 리스트나 튜플에 있는 인덱싱 방법을 적용할 수 없다.
```

```python
# 딕셔너리 만들 때 주의사항

# 딕셔너리에서 KEY는 고유한 값이므로 중복되는 KEY 값을 설정해 놓으면 하나를 제외한 나머지 것들이 모두 무시된다는 점에 주의해야 한다.

a = {1:'a', 1:'b'}
a

# 동일한 KEY 중 1개를 제외한 나머지 KEY: Value값이 모두 무시됨
# 딕셔너리에는 동일한  KEY가 중복으로 존재할 수 없다.
```

```python
# 딕셔너리 관련 함수

a = {'name': 'Pey', 'phone': '010-0000-1111', 'birth': '1118'}
a.keys()

# a.keys()는 딕셔너리a의 KEY만을 모아 dict_keys 객체를 반환한다.
```

```python
# dict_keys 객체는 다음과 같이 사용할 수 있음
# 리스트를 사용하는 것과 별 차이는 없지만, 리스트 고유의 append, insert, pop, remove, sort 함수는 수행할 수 없다.

for k in a.keys():
    print(k)
```

```python
# dict_keys 객체를 리스트로 변환하는 방법

list(a.keys())
```

```python
# value 리스트 만들기 - values
a.values()

# Value 값만 얻고 싶다면, values 함수를 사용하면 된다
# values 함수를 호출하면, dict_values 객체를 반환한다.
```

```python
# KEY, Value 쌍 얻기 - items

a.items()

# items 함수는 key와 value의 쌍을 튜플로 묶은 값을 dict_items객체로 반환한다.
```

```python
# KEY: Value 쌍 모두 지우기 - clear

a.clear()
a
```

```python
# KEY로 Value 얻기 - get

a = {'name': 'Pey', 'phone': '010-0000-1111', 'birth': '1118'}
print(a.get('name'))
print(a['name'])
```

```python
print(a.get('phone'))
print(a['phone'])

print(a.get('nokey'))
print(a['nokey'])
```

```python
# a.get('name')은 a['name']을 사용했을 때와 동일한 결괏값을 반환한다.

# a['nokey']처럼 딕셔너리에 존재하지 않는 키로 값을 가져오려고 할 경우, a['nokey']방식은 오류를 발생시키고, a.get('nokey')방식은 None을 반환한다는 차이가 있다.
```

```python
a.get('nokey', '정보없음')
```

```python
# 해당 KEY가 딕셔너리 안에 있는지 조사하기 - in

a = {'name': 'Pey', 'phone': '010-0000-1111', 'birth': '1118'}
'name' in a
```

```python
'email' in a
```

```python
t1 = (1,2,3,4)

4 in t1
```

```python
# KEY로 Value 얻기 - pop

a = {'name': 'Pey', 'phone': '010-0000-1111', 'birth': '1118'}

phone = a.pop('phone')
phone
```

```python
a
```

```python
#pop(X) 함수는 딕셔너리에서 KEY가 x 인 항목을 삭제한 후 그 Value 값을 반환한다. get 함수와 달리 pop은 해당 항목을 딕셔너리에서 삭제한다는 차이가 있다.
```
