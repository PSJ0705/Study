02-7 불 자료형

```python
# 불 자료형이랑 참(True)와 거짓(False)을 나타내는 자료형이다.
# 불 자료형은 다음 2가지 값만을 가질 수 있다.
```

```python
# 불 자료형 사용법

a = True
b = False
```

```python
print(type(a))
print(type(b))
```

```python
a = [1,2,3,4]
while a:
    print(a.pop())
```

```python
if []:
    print("참")
else:
    print("거짓")
```

```python
if [1,2,3]:     # [1,2,3] 은 요솟값이 있는 리스트이므로 무조건 참을 출력함
    print("참")
else:
    print("거짓")
```

```python
# 불 연산

bool('python')  # 'python' 문자열은 비어있지 않으므로 bool 연산의 결과로 불 자료형인 True를 반환함
```

```python
bool('')    #'' 문자열은 비어있으므로 bool 연산의 결과로 False를 반환함
```

```python
print(bool([1,2,3]))

print(bool([]))

print(bool(0))

print(bool(3))
```

논리 연산자

- 불 자료형에서는 and, or, not 과 같은 논리 연산자를 사용할 수 있다.

```python
# And 연산자
# 양쪽 조건이 모두 참일 때만 True를 반환한다.

print(True and True)

print(True and False)

print(False and True)

print(False and False)
```

```python
# Or 연산자
# 양쪽 조건 중 하나라도 참이면 True를 반환한다.

print(True or True)

print(True or False)

print(False or True)

print(False or False)
```

```python
# Not 연산자
# 조건의 참/거짓을 뒤바꾼다.

print(not True)

print(not False)

print(not 0)

print(not 1)

```

```python
x = 5
y = 10

print(x > 0 & y > 0)
print(x > 10 | y > 5)
print(not x > y)
```
