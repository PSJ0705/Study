# 03-3 for문

for 문의 기본 구조

    for 변수 in 리스트(또는 튜플, 문자열):
        수행할_문장1
        수행할_문장2


```python
# 전형적인 for문

test_list = ['one', 'two', 'three']

for i in test_list:
    print(i)
```

```python
# 또 다른 for문의 사용

a = {(1,2), (3,4), (5,6)}

for (first,last) in a:
    print(first + last)

# a 리스트의 요솟값이 튜플이기 때문에 각각의 요소가 자동으로 (first, last) 변수에 대입된다.
```

```python
# for 문의 응용

# 총 5명의 학생이 시험을 보았는데, 시험점수가 60점 이상이면 합격이고 그렇지 않다면 불합격이다.
# 합격인지, 불합격인지 결괄르 보여주시오.


marks = [95, 68, 40, 50, 70]        #학생들이 점수 리스트

number = 0      # 학생들에게 붙여 줄 번호

for mark in marks:      # 95, 68, 40, 50 ,70을 순서대로 mark에 대입
    number += 1
    if mark >= 60:
        print("%d번 학생은 합격입니다." % number)
    else:
        print("%d번 학생은 불합격입니다." % number)
```

```python
# for 문과 continue 문

#  for 문 안의 문장을 수행하는 도중 continue 문을 만나면 for문의 처음으로 돌아가게된다.
```

```python
# marks2. py

marks = [95, 68, 40, 50, 70]

number = 0

for mark in marks:
    number += 1
    if mark < 60:
        continue        # 점수가 60점 미만인 경우에는 continue 문이 수행되어 for 문의 처음으로 돌아간다.
    print("%d번 학생 축하합니다. 합격입니다."  % number)
```

```python
#  for 문과 함께 자주 사용하는 range 함수

# range 함수의 간단한 사용법

a = range(10)
a

# range(10)은 0부터 10미만의 숫자를 포함하는 range객체를 만들어준다.
# 시작 숫자와 끝 숫자를 포함하려면 range(시작_숫자, 끝_숫자) 형태를 사용하는데, 이 때 끝 숫자는 포함되지 않는다.
```

```python
# range 함수의 예시 살펴보기

add = 0

for i in range(1,11):
    add += i
    print(add)
```

```python
# range  함수를 사용한 marks3. py

marks = [95, 68, 40, 50, 70]

for number in range(len(marks)):
    if marks[number] <  60:
        continue
    print("%d번 학생 축하합니다. 합격입니다." % number)
```

```python
# for 문과 range 문을 활용한 구구단

for i in range(2,10):
    for j in range(1,10):
        print("%d X %d = %d" % (i,j,i*j))
    print("===============")
```

```python
# 리스트 컴프리헨션 사용하기

# 리스트 안에 for 문을 포함하는 리스트 컴프리헨션(List comprehension)을 사용하면 좀 더 편리하고 직관적인 프로그램을 만들 수 있다.

a = [1,2,3,4]
result = []

for num in a:
    result.append(num*3)
    print(result)
```

```python
# 위 코딩에 리스트 컴프리헨션을 사용한 코드

a = [1,2,3,4]

result = [num * 3 for num in a]
print(result)
```

```python
# [1,2,3,4] 중에서 짝수에만 3을 곱하여 담는 코드

a = [1,2,3,4]

result = [num * 3 for num in a if num % 2 == 0]
print(result)
```

```python
# 리스트 컴프리헨션의 문법

# [포현식 for 항목 in 반복_가능_객체 if 조건문]

# [포현식 for 항목1 in 반복_가능_객체1 if 조건문1
#        for 항목2 in 반복_가능_객체2 if 조건문2
#        ...
#        for 항목n in 반복_가능_객체n if 조건문n]
```

```python
# 리스트 컴프리헨션을 활용한 구구단의 모든 결과

result = [x * y for x in range(2,10)
                for y in range(1,10)]
print(result)
```

```python
# for 문과 break 문

# while 문에서와 마찬가지로 for 문에서도 break문을 사용할 수 있다. break 문은 for 문을 강제로 빠져나가고 싶을 때 사용한다.

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# for-else문

# 파이썬의 for문에는 else절을 사용할 수 있다. for-else문은 for문이 정상적으로 종료되었을 때(break로 빠져나가지 않았을 때) else절이 실행된다.

for i in range(5):
    print(i)
else:
    print("for문이 정상 종료되었습니다.")
```

```python
# enumerate 함수 활용하기

# 리스트의 순서(인덱스)와 값을 함께 구하고 싶을 때에는 enumerate 함수를 사용하면 편리하다.

fruits = ['apples', 'banana', 'oranges']
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}: {fruit}")
```

```python
# zip함수로 여러 리스트 함께 순회하기

# 두 개 이상의 리스트를 동시에 순회하고 싶을 때는 zip 함수를 사용한다.

names = ['홍길동', '김철수','이영희']
scores = [85,90,75]

for name, score in zip(names, scores):
    print(f"{name}: {score}")
```

```python
# 세 개의 리스트 함께 순회하기

names = ['철수', '영희', '맹구']
korean = [85,90,75]
math = [85,70, 95]

for name, kor, mat in zip(names, korean, math):
    print(f"{name}: 국어 {kor}점, 수학 {mat}점")
```
