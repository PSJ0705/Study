## 04-4 프로그램의 입출력

명령 프롬프트의 명령어 중 type a.txt

type 은 바로 뒤에 적힌 파일 이름을 인수로 받아 해당 파일 내용을 출력해 주는 명령어다.

        명령어 [인수1 인수2...]

```python
# sys 모듈 사용하기
# sys 모듈을 사용하려면 import sys 처럼 import 명령어를 사용해야 한다.

import sys

args = sys.argv[1:]

for i in args:
    print(i)
```

위 프로그램 실행 시 전달받은 인수를 for문을 사용해 차례대로 하나씩 출력하는 예이다. sys 모듈의 argv는 프로그램 실행 시 전달된 인수를 의미한다.

```python
import sys
args = sys.argv[1:]
for i in args:
    print(i.upper(), end=' ')
```
