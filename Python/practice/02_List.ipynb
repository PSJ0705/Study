리스트 관련 함수

```python
#리스트에 요소 추가하가 - append(첨부하다, 덧붙이다)

a = [1,2,3]
a.append(4)
a
```

```python
a.append([5,6])
a
```

```python
#리스트 정렬 - sort

a = [1,5,3,2,4]
a.sort()
a
```

```python
a = ['b', 'c', 'a', 'd']
a.sort()
a
```

```python
#리스트 뒤집기 - reverse

a = ['a', 'c', 'b', 'e']
a.reverse()
a
```

```python
# 인덱스 반환 - index

a = [1,2,3]
a.index(1)
```

```python
# 리스트에 요소 삽입 - insert

a = [1,2,3]
a.insert(0,4)       #0번째 자리, 즉 a[0] 위치에 4를 삽입하라는 뜻
a
```

```python
a.insert(3,5)
a
```

```python
#리스트 요소 제거 - remove

a = [1,2,3,1,2,3]
a.remove(3)
a

#첫 번째 3만 제거됨
```

```python
a.remove(3)
a

#한 번 더 실행하면 다시 3이 삭제됨
```

```python
# 리스트 요소 끄집어 내기 - pop

a = [1,2,3]
a.pop()
a

#a 리스트 [1,2,3]에서 3을 끄집어내고 최종적으로 [1,2]만 남는다
#pop(x)는 리스트의 x 번째 요소를 반환하고 그 요소는 삭제한다.
```

```python
a = [1,2,3]
a.pop(1)
a

#a.pop(1) a[1]의 값을 끄집어내어 반환한다.
```

```python
#리스트에 포함된 요소 x의 개수 세기 - count

a = [1,2,3,4]
a.count(1)

#1이라는 값이 리스트 a에 2개 들어가있으므로 2를 반환한다.
```

```python
a = ['a', 'A', 'B', 'b', 'C', 'A']
a.count('a')

#대소문자 구별함
```

```python
#리스트 확장 - extend

a = [1,2,3]
a.extend([4,5])
a
```

```python
b = [6,7]
a.extend(b)
a

# a.extend([4,5])는 a += [4,5]와 동일하다
```
