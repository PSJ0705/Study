## 04-3 파일 읽고 쓰기

```python
# 파일 생성하기

f = open("새파일.txt", 'w')
f.close()
```

파일을 생성하기 위해 파이썬 내장함수 open을 사용했다.

open 함수는 다음과 같이 '파일이름'과 '파일 열기 모드'를 입력값으로 받고, 결괏값으로 파일 객체를 반환한다.

파일 열기 모드
- r : 읽기모드, 파일을 읽기만 할때 사용한다.
- w : 쓰기모드, 파일에 내용을 쓸 때 사용한다.
- x : 추가모드, 파일의 마지막에 새로운 내용을 추가할 때 사용한다.


파일을 쓰기 모드로 열면, 해당 파일이 이미 존재할 경우 원래 있던 내용이 모두 사라지고 해당 파일이 존재하지 않으면 새로운 파일이 생성된다.

```python
# '새파일.txt'파일을 다른 디렉터리에 생성하고 싶다면 다음과 같이 작성해야 한다.

f = open("C:\박상준\새파일.txt", 'w')
f.close()

# 이 모드에서 f.close()는 열려 있는 파일 객체를 닫아주는 역할을 함
```

```python
# 파일을 쓰기 모드로 열어 내용쓰기

f = open("C:\박상준\새파일.txt", 'w')

for i in range(1,11):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

```python
for i in range(1,11):
    data = "%d번째 줄입니다." % i
    print(data)
```

### 파일을 읽는 여러가지 방법

```python
# 1. readline 함수 이용하기

f = open("C:\박상준\새파일.txt", 'r')

line = f.readline()
print(line)
f.close()

# 이 코드는 '새파일.txt'파일을 읽기 모드로 연 후 readline()을 사용해서 파일의 첫 번째 줄을 읽어 출력하는 예제이다.
```

```python
# 모든 줄을 읽어 화면에 출력하는 코드

f = open("C:\박상준\새파일.txt", 'r')

while True:
    line = f.readline()
    if not line : break
    print(line)
f.close()
```

```python
# 2. readlines 함수 사용하기

f = open("C:\박상준\새파일.txt", 'r')

lines = f.readlines()

for line in lines:
    print(line)
f.close()

# readlines 함수는 파일의 모든 줄을 읽어서 각각의 줄을 요소로 가지는 리스트를 반환한다. 위 예에서 lines는 ["1번째 줄입니다.","2번째 줄입니다.","3번째 줄입니다.",...]가 된다.
```

```python
# 줄바꿈(\r)문자 제거하기

f = open("C:\박상준\새파일.txt", 'r')
lines = f.readlines()

for line in lines:
    line = line.strip()         # 줄끝의 줄 바꿈 문자를 제거한다.
    print(line)
f.close()
```

```python
# 3. read 함수 사용하기

f = open("C:\박상준\새파일.txt", 'r')
data = f.read()
print(data)
f.close()

# f.read()는 파일의 내용 전체를 문자열로 반환한다.
```

```python
# 4. 파일 객체를 for 문과 함께 사용하기

f = open("C:\박상준\새파일.txt", 'r')

for line in f:
    print(line)
f.close()
```

### 파일에 새로운 내용 추가하기

쓰기 모드('w')로 파일을 열 때 이미 존재한느 파일을 열면 그 파일의 내용이 모두 사라진다.

원래 있던 값을 유지하면서 단지 새로운 값만 추가해야 할 경우, 파일을 추가모드('a')로 열면 된다.

```python
f = open("C:\박상준\새파일.txt", 'a')

for i in range(11,31):
    data = "%d 번째 줄입니다.\n" %i
    f.write(data)
f.close()
```

### with 문과 함께 사용하기

```python
with open("C:\박상준\새파일.txt", "w") as f:
    f.write("Life is too short. you need Python")

# 다음과 같이 with 문을 사용하면 with 블록(with문에 속해있는 문장)을 벗어나는 순간, 열린 파일 객체 f가 자동으로 닫힌다.
```

### with 문과 스코프(scope) 규칙

- 파이썬에서 with 문 내에서 선언된 변수는 with 블록을 벗어난 후에도 접근할 수 있다. 이는 파이썬의 스코프(유효범위) 특성 때문이다.

```python
# 1. 함수 스코프 : 함수 안에서 선언된 변수는 함수 밖에서 접근할 수 없다.

def my_function():
    func_var = "함수 안의 변수"

my_function()
print(my_function())
# print(func_var_)      # 오류! 함수 밖에서는 접근 불가
```

```python
# 2. 블록 스코프 : if, for, while, with 등의 블록 안에서 선언된 변수는 블록 밖에서도 접근할 수 있다.

# if 문의 예시
if True:
    if_var = "if 블록 안의 변수"

print(if_var)

# for 문 블록의 예시

for i in range(3):
    loop_var = "반복문 안의 변수"

print(i)
print(loop_var)

# with 문에서 변수 사용 예지

with open("test.txt", "w") as f:
    content = "Hello, Python!"
    f.write(content)

print(content)
```

### 파일 처리시 주의사항

한글이 포함된 파일을 다룰 때는 인코딩을 명시하는 것이 좋다.

```python
# 한글 파일 쓰기

with open("한글파일.txt", "w", encoding="utf-8") as f:
    f.write("안녕하세요, 파이썬!")



```

```python
# 한글 파일 읽기

with open("한글파일.txt", "r", encoding="utf-8") as f:
    content = f.read()
    print(content)
```
