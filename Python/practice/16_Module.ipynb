## 모듈 불러오기

import의 사용 방법 : import 모듈_이름 (모듈 이름은 확장자를 제거한 파일 이름을 가리킴)

모듈 이름 없이 함수 이름만 쓰는 방법 : from 모듈_이름 import 모듈_함수

Ex) from mod1 import add, sub

Ex) from mod1 import *

## if __name__ == "__main__":의 의미

if __name__ == "__main__"을 사용하면 C:\doit>python mod1.py처럼 직접 이 파일을 실행했을 때는 __name__ == "__main__"이 참이 되어 if 문 다음 문장이 수행된다. 반대로 대화형 인터프리터나 다른 파일에서 이 모듈을 불러 사용할 때는 __name__ == "__main__"이 거짓이 되어 if 문 다음 문장이 수행되지 않는다.

### __name__ 변수란?

파이썬의 __name__ 변수는 파이썬이 내부적으로 사용하는 특별한 변수 이름이다. 파일을 직접 실행할 경우 __name__ 변수에는 __ main__값이 실행된다. 하지만 파이썬 모듈에서 import 할 경우에는 __name__ 변수에는 모듈 이름이 저장된다.
