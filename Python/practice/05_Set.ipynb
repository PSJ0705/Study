집합 자료형

- 집합(set)은 수학의 집합 개념과 동일한 자료형
- 중복은 허용하지 않고, 순서가 없는 데이터들의 모임
- 교집합, 합집합, 차집합 등의 집합 연상을 쉽게 처리할 수 있음

```python


# 집합 자료형 만드는 방법

s1 = set([1,2,3])
s1
```

```python
s2 = set("Hello World")
s2
```

```python
s3 = {1,2,3}
s3
```

```python
s4 = {'a', 'b', 'c'}
s4

# 비어있는 집합 자료형은 s = set()로 만들 수 있다. s = {}로 만들면 딕셔너리가 되어버리니 주의할 것
```

```python
# 중간 정리

a = [1,2,3] #리스트 자료형
a = (1,2,3) #튜플 자료형
a = {'K': 'V'} #딕셔너리 자료형
a = set([1,2,3]) #집합 자료형1
a = set("Hello") #집합 자료형2
a = {1,2,3} #집합 자료형3

a
```

```python
#집합 자료형의 특징
# 1. 중복을 허용하지 않는다.
# 2. 순서가 없다.

# => 집합은 중복을 허용하지 않는 특징 때문에 데이터의 중복을 제거하기 위한 필터로 종종 사용된다.
```

```python
a = {3,1,2, 100,50,218,3800,10}
print(a)
```

```python
for i in range(5):
    print(a)
```

```python
# 집합 자료형은 순서가 없기 때문에 인덱싱을 통해 요솟값을 얻을 수 없다.
# 만약 집합 자료형에 저장된 값을 인덱싱으로 접근하려면 다음과 같이 리스트나 튜플로 변환한 후에 해야 한다.
```

```python
s1 = set([1,2,3])
l1 = list(s1) #s1을 리스트로 변환
print(l1)
print(l1[0])

t1 = tuple(s1) #s1을 튜플로 변환
print(t1)
print(t1[0])
```

```python
# 교집합, 합집합, 차집합 구하기
# 집합 자료형을 정말 유용하게 사용하는 경우는 교집합, 합집합, 차집합을 구할 때이다.

s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
```

```python
# 교집합 구하기

print(s1 & s2)

s1.intersection(s2) #동일하게 교집합 구하기
```

```python
# 합집합 구하기

print(s1|s2)

print(s1.union(s2)) #합집합 구하기 함수
```

```python
# 차집합 구하기

print(s1-s2)
print(s2-s1)

print(s1.difference(s2))  #차집합 구하기 함수
print(s2.difference(s1))  #차집합 구하기 함수
```

```python
# 집합 자료형 관련 함수

# 값 1개 추가하기 - add

s1 = set([1,2,3])
s1.add(4)
print(s1)
```

```python

# 값 여러 개 추가하기 - update

s1 = set([1,2,3])
s1.update([4,5,6])
print(s1)
```

```python

# 특정 값 제거하기 - remove

s1 = set([0,1,2,3,4,5,6])

for i in range(6):
    s1.remove(i)
    print(s1)
```

```python

# 특정 값 제거하기 - discard

s1 = set([1,2,3])
s1.discard(2)
print(s1)

s1.discard(4)   # 없는 값이어도 오류가 발생하지 않음
print(s1)
```

```python

# 모든 값 제거하기 - discard

s1 = set([1,2,3])
s1.clear()
print(s1)
```
