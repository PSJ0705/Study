```python
# 변수란?
# 파이썬에서 사용하는 변수는, 객체를 가리키기 위한 것
```

```python
a = [1,2,3]
b = a
# [1,2,3]이라는 리스트 객체를 참조하는 변수가 a 변수 1개에서 b 변수가 추가되어 2개로 늘어났다는 차이가 존재
```

```python
print(id(a))

print(id(b))    # id 값이 동일함 = a가 가리키는 대상과 b가 가리키는 대상이 동일함
```

```python
b.append(4)
print(a)
print(b)
```

```python
a[1] = 4
print(a)
print(b)
```

```python
# b 변수를 가져올 때, a 변수의 값을 가져오면서 a 와는 다른 주소를 가리키도록 하는 방법 2가지
```

```python
# 1. [:] 이용하기
# 리스트 전체를 가리키는 [:]을 사용해서 복사하는 것

a = [1,2,3]
b = a[:]
a[1] = 4
print(a)
print(b)

print(id(a))
print(id(b))
```

```python
# 2. copy 모듈 이용하기

from copy import copy

a = [1,2,3]
b = copy(a) # b = copy(a)는 b = a[:}와 동일하다
# 혹은 b = a.copy() 함수를 사용해도 동일한 결과를 얻을 수 있다.
print(a is b)
```
