02-4 튜플 자료형

```python
# 리스트는 [], 튜플은 ()으로 둘러싸인다.
# 리스트는 요솟값의 생성, 삭제, 수정이 가능함
# 튜플은 요솟값을 바꿀 수 없다.
```

```python
#튜플의 여러가지 모습

t1 = ()
t2 = (1,)
t3 = (1,2,3)
t4 = 1,2,3
t5 = ('a','b','c', 'd', 'e')

a = 1,2,3

#t2 = (1,) 처럼 1개의 요소만을 가질 때는 요소 뒤에 쉼표(,)를 붙여야 한다.
# t4 = 1,2,3,4처럼 소괄호()를 생략해도 된다.
```

```python
# 튜플과 리스트의 가장 큰 차이는, 요솟값을 변화시킬 수 있는지의 여부
# 리스트의 요솟값은 변화가 가능하고, 튜플의 요솟값은 변화가 불가능하다.
```

튜플 다루기

```python
#인덱싱하기

t1 = (1,2,'a','b')
print(t1[3])
t1[3]
```

```python
# 슬라이싱하기

t1 = (1,2,'a','b')
t1[1:]
```

```python
# 튜플 더하기

t1 = (1,2,'a','b')
t2 = (3,4)
t3 = t1 + t2
print(t3)
```

```python
# 튜플 곱하기

t2 = (3,4)
t3 = t2 * 3
t3
```

```python
# 튜플 길이 구하기
t1 = (1,2,'a','b')
len(t1)
```

```python
# 튜플은 요솟값을 변경할 수 없기 때문에 sort, insert, remove, pop과 같은 내장함수가 없다
```
