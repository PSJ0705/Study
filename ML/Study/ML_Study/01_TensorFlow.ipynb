## TensorFlow Hello World!

```python
# create a constant op
# this op is added as a node to the default graph

import tensorflow as tf
```

```python

hello = tf.constant("Hello, TensorFlow!")

# run the op and get result
print(hello.numpy())
```

## Computational Graph

```python
node1 = tf.constant(3.0, tf.float32)
node2 = tf.constant(4.0)
node3 = tf.add(node1, node2)
```

```python
print("node1:", node1, "node2:", node2)
print("node3:", node3)
```

```python
print("node1 :", node1.numpy(), "node2: ", node2.numpy())
print("node3: ", node3.numpy())
```

## 값 입력받아서 처리

```python
a = float(input("a 값을 입력하세요 : "))
b = float(input("b 값을 입력하세요 : "))

# Tensor로 변환
a = tf.constant(a, dtype=tf.float32)
b = tf.constant(b, dtype=tf.float32)

# 연산 정의
adder_node = a + b

# 결과 출력
print("결과 : ", adder_node)
```

TF가 내부적으로 그래프로 만들어서 최적화해서 돌려줌

```python
@tf.function
def adder(a,b):
    return a + b

a = float(input("a 값을 입력하세요 : "))
b = float(input("b 값을 입력하세요 : "))

result = adder(a,b)
print(result.numpy())
```

위 방법으로는 a에 [1,3, 4... ]처럼 여러 개의 값을 한번에 넣을 수 없음

 벡터/리스트 활용

```python

# 여러 값을 한 번에 입력받기(공백구분)
a_str = input("a 값을 공백으로 입력( 예: 1 2 3 ) : ")
b_str = input("b 값을 공백으로 입력( 예: 4 5 6 ) : ")

# 문자열을 float 리스트로 변환
a_list = list(map(float, a_str.split()))
b_list = list(map(float, b_str.split()))

# 텐서로 변환
a = tf.constant(a_list, dtype=tf.float32)
b = tf.constant(b_list, dtype=tf.float32)

result = adder(a,b)
print(result.numpy())
```
