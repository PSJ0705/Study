# Multi-variable linear regression

hypothesis = H(x) = Wx + b

=> H(x1, x2, x3) => w1x1 + w2x2 + w3x3 + b

### Matrix multiplication(메트릭스 곱셈)

Hypothesis using matrix

(x1, x2, x3) * (w1  = (w1x1 + w2x2 + w3x3)
                w2
                w3)

### WX vs XW

- Lecture(theory) : H(x) = Wx + b
- Implementation(TensorFlow) : H(X) = XW
